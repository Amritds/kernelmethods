```python
a= np.percentile(theta1,5)
b =np.percentile(theta1,50)
c =np.percentile(theta1,75)
d = np.percentile(theta1,95)
```

```python
import matplotlib.pyplot as plt
plt.clf()
fig=plt.figure()
plt.axhline(y=a,color ='r')
plt.axhline(y=b,color ='g')
plt.axhline(y=c,color ='b')
plt.axhline(y=d,color ='y')
plt.legend(['5th percentile','50th percentile','75th percentile','95th percentile'])
plt.show()
fig.savefig('ARCHPolyAutoPercentiles_para1.pdf')
```
