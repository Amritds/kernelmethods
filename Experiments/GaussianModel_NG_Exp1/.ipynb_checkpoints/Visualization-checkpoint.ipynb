```python
import numpy as np
kD =np.load('klDivergence.npy')
```

```python
a= np.percentile(kD,5)
b =np.percentile(kD,50)
c =np.percentile(kD,75)
d = np.percentile(kD,95)
```

```python
import matplotlib.pyplot as plt
plt.clf()
fig=plt.figure()
plt.axhline(y=a,color ='r')
plt.axhline(y=b,color ='g')
plt.axhline(y=c,color ='b')
plt.axhline(y=d,color ='y')
plt.legend(['5th percentile','50th percentile','75th percentile','95th percentile'])
plt.show()
fig.savefig('GaussianRBFAutoPercentiles.pdf')
```

```python
import matplotlib.pyplot as plt
from scipy.stats import gaussian_kde
density = gaussian_kde(kD)
xs = np.linspace(-2,2,50)
plt.plot(xs,density(xs))
plt.show()
```
