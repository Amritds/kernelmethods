```python
import pickle
import numpy as np
import Utilities.kernelClassifier as kC
import Utilities.hyperparameterChoice as hC
from Utilities.ExpModels import  GaussianModel
from Utilities.kernels import kernelLinear, kernelPoly, kernelRBF

#Random Seed
randomSeed = 432

#------Exp Setup -------------------------------------------------#
# Model
model = GaussianModel(3,-20,20, randomSeed) 
typicalThetaValue = 0
#Number of samples to be created by each generator for classification.
n_samples=1000
```

```python
#----Select parameters -------------------------------------------#
#Select parameter values and reg scale for the polynomial kernel.
kPoly =lambda X1,X2: kernelPoly(X1,X2,2) #Any setting is fine.
clf=kC.KernelClassifier(kPoly)
regScale1 = hC.autoRegScale(clf,model,typicalThetaValue ,n_samples, numberOfOptions =20)

```

```python
#----Select parameters -------------------------------------------#
#Select parameter values and reg scale for the polynomial kernel.
kPoly =lambda X1,X2: kernelPoly(X1,X2,4) #Any setting is fine.
clf=kC.KernelClassifier(kPoly)
regScale2 = hC.autoRegScale(clf,model,typicalThetaValue ,n_samples, numberOfOptions =20)
```

```python
import pickle
with open('ManualExps_parametersAndRegScales.pickle', 'rb') as f:
        [(g1,regScale1),(g2,regScale2)] = pickle.load(f)
```

```python
observedData = np.array([[0]])
# Approximate posterior for values 
mu_values=np.linspace(-2*3+ observedData[0,0],2*3+ observedData[0,0] ,100)

kPoly1 =lambda X1,X2: kernelPoly(X1,X2,2) #Any setting is fine.
clf1=kC.KernelClassifier(kPoly1)
kernelsAndRegScales = [(clf1,regScale1)]
```

```python
from Utilities.LFIRE_Estimator import PosteriorEstimator
#--------Run the ratio estimation algorithm ----------------------------------------------------------------------------
Estimator = PosteriorEstimator(randomSeed)
probabilities = Estimator.estimateProbabilities(observedData, mu_values, model, n_samples, kernelsAndRegScales)
with open('ManualExps_poly2Probabilities.pickle', 'wb') as f:
    pickle.dump(probabilities, f)
```

```python
#-------True probabilites are available for the Gaussian model, plot alongside approximations -------------------------
import matplotlib.pyplot as plt
plt.clf()
fig = plt.figure()
ax = fig.add_subplot(1,1,1)
ax.plot(mu_values,model.truePosterior(mu_values,observedData),'g-')
ax.plot(mu_values,probabilities[0],'b-')
plt.show()
fig.savefig('ObservedData_0_Polydegree2_kernel_GaussianModel.pdf')
```

```python
kPoly2 =lambda X1,X2: kernelPoly(X1,X2,4) #Any setting is fine.
clf2=kC.KernelClassifier(kPoly2)
kernelsAndRegScales = [(clf2,regScale2)]

from Utilities.LFIRE_Estimator import PosteriorEstimator
#--------Run the ratio estimation algorithm ----------------------------------------------------------------------------
Estimator = PosteriorEstimator(randomSeed)
probabilities = Estimator.estimateProbabilities(observedData, mu_values, model, n_samples, kernelsAndRegScales)
with open('ManualExps_poly4Probabilities.pickle', 'wb') as f:
    pickle.dump(probabilities, f)
#-------True probabilites are available for the Gaussian model, plot alongside approximations -------------------------
import matplotlib.pyplot as plt
plt.clf()
fig = plt.figure()
ax = fig.add_subplot(1,1,1)
ax.plot(mu_values,model.truePosterior(mu_values,observedData),'g-')
ax.plot(mu_values,probabilities[0],'b-')
plt.show()
fig.savefig('ObservedData_0_Polydegree4_kernel_GaussianModel.pdf')
```
